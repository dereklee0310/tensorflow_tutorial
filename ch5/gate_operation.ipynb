```python
# single qubit gate
print(cirq.H)
print(np.real(cirq.unitary(cirq.H)))

# two-qubit gate
print(cirq.CNOT)
print(np.real(cirq.unitary(cirq.CNOT)))

print(cirq.MeasurementGate)
```

**Gate to operation**

```python
# gate
print(cirq.H, type(cirq.H).__name__)

# operation
qubit = cirq.NamedQubit('qubit1')
hadamard_operation = cirq.H(qubit)
print(hadamard_operation, type(hadamard_operation).__name__)
```
