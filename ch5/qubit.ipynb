**Create different kinds of qubits**

```python
named_qubit = cirq.NamedQubit('myqubit')
print(named_qubit)

line_qubit = cirq.LineQubit(1)
print(line_qubit)

grid_qubit = cirq.GridQubit(2, 2)
print(grid_qubit)
```

**Create multiple qubits at once**

```python
# list of LineQubits
line_qubits = cirq.LineQubit.range(4)
print(line_qubits)

# square of GridQubits
square_qubits = cirq.GridQubit.square(2, 2)
print(square_qubits)

# rectangle of GridQubits
rect_qubits = cirq.GridQubit.rect(2, 3, 0, 1)
print(rect_qubits)
```
