```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))

x = np.arange(-10, 11, 0.1)
y = sigmoid(x)
plt.title('Sigmoid')
plt.axvline(0, c='k', linewidth='1')
plt.axhline(0, c='k', linewidth='1')
plt.plot(x, y, linewidth='3')
plt.show()
```

```python
tanh = lambda x: (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
y = tanh(x)
plt.title('Tanh')
plt.axvline(0, c='k', linewidth='1')
plt.axhline(0, c='k', linewidth='1')
plt.plot(x, y, linewidth='3')
plt.show()
```

```python
relu = lambda x : [max(x_, 0) for x_ in x]
y = relu(x)
plt.title('ReLU')
plt.axvline(0, c='k', linewidth='1')
plt.axhline(0, c='k', linewidth='1')
plt.plot(x, y, linewidth='3')
plt.show()
```

```python
leaky_relu = lambda x, a : [max(x_, a * x_) for x_ in x]
y = leaky_relu(x, 0.1)
plt.title('Leaky ReLU')
plt.axvline(0, c='k', linewidth='1')
plt.axhline(0, c='k', linewidth='1')
plt.plot(x, y, linewidth='3')
plt.show()
```
