```python
import tensorflow as tf
```

**Create a variable**

```python
variable = tf.Variable([0.2 ,0.4], name='var1')
print(variable)
```

**Modify a variable**

```python
tf.print("before:", variable)
variable[0].assign(0.6)
variable[1].assign(0.8)
tf.print("after:", variable)
```

**Convert it to tensor**

```python
print("before:", variable)
print("after:", tf.convert_to_tensor(variable))
```

**Create a non-trainable variable**

```python
variable1 = tf.Variable([23, 95], trainable=False, name='another_var')
print(variable1)
```
