**Create tensors**

```python
import tensorflow as tf

scalar = tf.constant(5)
rank1_tensor = tf.constant([1, 2, 3])

tensor_3x1 = tf.constant([[1],
                          [2],
                          [3]])

tensor_3x2 = tf.constant([[1, 2],
                          [3, 4],
                          [5, 6]])
```

**Scalar broadcasting**

```python
tf.print("add 5 to each elements:", tf.add(rank1_tensor, scalar))
tf.print("subtract 5 from each elements:", tf.subtract(rank1_tensor, scalar))
tf.print("multiply each elements by 5:", tf.multiply(rank1_tensor, scalar))
tf.print("divider each elements by 5:", tf.divide(rank1_tensor, scalar))
```

**Tensor Broadcasting**

```python
tf.print("add up two tensors row by row:\n", tf.add(tensor_3x1, tensor_3x2))
```
