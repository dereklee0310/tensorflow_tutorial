**Migrate to TensorFlow 2**

```python
import tensorflow as tf

@tf.function
def graph(a, b):
    c = a + b
    return c

tf.print("a + b =", graph(1, 1))
```

**Placeholder in TensorFlow 1.x**

```python
import tensorflow.compat.v1 as tf1
tf1.disable_eager_execution()

a = tf1.placeholder(dtype=tf.int32)
b = tf1.placeholder(dtype=tf.int32)
c = a + b

with tf1.Session() as sess:
  print("a + b =", sess.run(c, feed_dict={a: 1, b: 1}))
```
