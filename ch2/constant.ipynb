**Import TensorFlow**

```python
import tensorflow as tf
```

**Create constants**

```python
string = tf.constant('John', tf.string)
integer = tf.constant(18, dtype=tf.int16, name='int')
floating = tf.constant(65.3, tf.float32)
```

**Data type can be omitted**

```python
number = tf.constant(81.274)
```

**Two types of printing**

```python
print(string)
tf.print(integer)
```

**N-dimensional tensor**

```python
matrix = tf.constant([[1, 2], [3, 4], [5, 6]], shape=(3, 2))
tf.print(matrix)
tf.print(matrix.dtype)
tf.print(matrix.shape)
```

**Access element**

```python
tf.print(matrix[1][0])
tf.print(matrix[2][1])
```

**Unmatched input**

```python
tf.print(integer + float) # 18 + 65.3 = 83.3
```

**Type casting**

```python

tf.print(tf.cast(integer, tf.float32) + float) # 18 + 65.3 = 83.3
```
