**Graph execution in TensorFlow 2.x**

```python
import tensorflow as tf

@tf.function
def graph():
    a = tf.constant(1)
    b = tf.constant(1)
    c = a + b
    return c
```

**Execute the graph**

```python
tf.print("a + b =", graph())
```

**Create graph in TensorFlow 1.x**

```python
import tensorflow.compat.v1 as tf1
tf1.disable_eager_execution()

a = tf.constant(1)
b = tf.constant(1)
c = a + b
```

**Execute the graph**

```python
with tf1.Session() as sess:
  print("a + b =", sess.run(c))
```
