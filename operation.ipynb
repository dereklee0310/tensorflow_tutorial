**Create tensor**

```python
import tensorflow as tf

a = tf.constant([1, 2, 3])
b = tf.constant([4, 5, 6])


# for matrix multiplication
a_3x2 = tf.constant([
                  [1, 2],
                  [3, 4],
                  [5, 6]])

b_2x3 = tf.constant([
                  [1, 2, 3],
                  [4, 5, 6]
                  ])
```

**Addition**

```python
c = tf.add(a, b)
tf.print("a + b =", c)
```

**Subtraction**

```python
c = tf.subtract(a, b)
tf.print("a - b =", c)
```

**Multiplication**

```python
c = tf.multiply(a, b)
tf.print("a * b =", c)
```

**Division**

```python
c = tf.divide(a, b)
tf.print("a / b =", c)
```

**Matrix multiplication**

```python
d = tf.matmul(a_3x2, b_2x3)
tf.print("[a] * [b] =", d)
```

**Transpose of matrix**

```python
tf.print("before:", a_3x2)
a_2x3 = tf.transpose(a_3x2)
tf.print("before:", a_2x3)
```

**Determinant of matrix**

```python
square = tf.constant([
                  [1, 2, 3],
                  [0, 4, 5],
                  [0, 0, 6]
                  ], dtype=tf.float32)

det = tf.linalg.det(square)
tf.print("determinant: ", det)
```
