**Import libraries**

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
```

**A helper function to plot the graphs**

```python
def plotData(x, y, title):
  plt.scatter(x, y)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(title)
  plt.show()
```


**Generate training data**

```python
x = np.linspace(0, 64, 256)
y = [1 if x_ > 48 else 0 for x_ in x]
plotData(x, y, "Training Data")
```

**Train model**

```python
x_reshaped = x[:, np.newaxis]
model = LogisticRegression()
model.fit(x_reshaped, y)

print("weight:", model.coef_[0])
print("intercept:", model.intercept_)
```

**Visualize result**

```python
linear_input = model.intercept_ + model.coef_[0] * x_reshaped
sigmoid = 1 / (1 + np.exp(-(linear_input)))
plt.plot(x_reshaped, sigmoid, 'r')
plotData(x, y, "Result")
```

**Sigmoid function**

```python
import matplotlib.pyplot as plt

map = lambda x: 1 / (1 + np.exp(-x))

X = np.linspace(-4, 4, 1024)
Y = [map(x_) for x_ in X]
plt.plot(X, Y)
plt.xlim(-4,4)
plt.ylim(0,1)
plt.grid()
plt.show()
```
